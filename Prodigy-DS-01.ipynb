# importing some library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## read file and show first 5 rows

```python
df = pd.read_excel('WorldBank Data.xlsx')
df.head()
```

## drop rows with empty columns

```python
df = df.dropna()
df.head()
```

# Internet

```python
# Group the data by 'Region' and calculate Internet usage
internet_usage_by_region = df.groupby('Region')['Individuals using the Internet (% of population)'].max().reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Individuals using the Internet (% of population)', data=internet_usage_by_region, palette='viridis')
plt.title('Internet Usage Across Regions')
plt.ylabel('Internet Usage (% of Population)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

```python
# Group the data by 'Year' and calculate Internet usage
life_expectancy_by_year = df.groupby('Year')['Individuals using the Internet (% of population)'].max().reset_index()

# Plotting the line chart
plt.figure(figsize=(20, 6))
plt.plot(life_expectancy_by_year['Year'], life_expectancy_by_year['Individuals using the Internet (% of population)'], marker='o', linestyle='-')

plt.title('Individuals using the Internet (% of population) per Year')
plt.xlabel('Year')
plt.ylabel('Individuals using the Internet (% of population)')
plt.grid(True)

plt.show()
```

## life expectancy at birth

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Life expectancy at birth (years)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Life Expectancy at Birth')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')

plt.show()
```

```python
# Group the data by 'Year' and calculate the mean of life expectancy
life_expectancy_by_year = df.groupby('Year')['Life expectancy at birth (years)'].max().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(life_expectancy_by_year['Year'], life_expectancy_by_year['Life expectancy at birth (years)'], marker='o', linestyle='-')

plt.title('Life Expectancy Over the Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.grid(True)

plt.show()
```

```python
sns.set(style="whitegrid")

# Group the data by 'Region' and calculate the mean of life expectancy
life_expectancy_by_region = df.groupby('Region')['Life expectancy at birth (years)'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Life expectancy at birth (years)', y='Region', data=life_expectancy_by_region, palette='viridis')

plt.title('Life Expectancy Across Regions')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Region')

plt.show()
```

## GDP (USD)

```python
# Group the data by 'Region' and calculate the max GPD
internet_usage_by_region = df.groupby('Region')['GDP (USD)'].max().reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='GDP (USD)', data=internet_usage_by_region, palette='viridis')
plt.title('GDP (USD) Usage Across Regions')
plt.ylabel('GDP (USD)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

## GDP per Capita (USD)

```python
# Group the data by 'Region' and calculate the max GPD per Capita
internet_usage_by_region = df.groupby('Region')['GDP per capita (USD)'].max().reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='GDP per capita (USD)', data=internet_usage_by_region, palette='viridis')
plt.title('GDP per capita (USD) Usage Across Regions')
plt.ylabel('GDP per capita (USD)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

## income Group

```python
# Group the data by 'Region' and count the incomeGroups
income_group_counts = df.groupby(['Region', 'IncomeGroup']).size().unstack(fill_value=0)

# Plotting the grouped bar chart
plt.figure(figsize=(14, 8))
income_group_counts.plot(kind='bar', stacked=True, colormap='viridis')

plt.title('Income Group Distribution Across Regions')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Income Group')

plt.show()
```

## Birth Rate and Death Rate

```python
# Group the data by 'Region' and compare the mean Birth Rate and Death Rate
birth_death_rates_by_region = df.groupby('Region')[['Birth rate, crude (per 1,000 people)', 'Death rate, crude (per 1,000 people)']].mean().reset_index()

# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Region', y='value', hue='variable', data=pd.melt(birth_death_rates_by_region, id_vars='Region'), palette='muted')

plt.title('Comparison of Birth and Death Rates Across Regions')
plt.ylabel('Rate (per 1,000 people)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Rate Type', loc='upper right')

plt.show()
```

```python
# Group the data by 'Year' and calculate the mean of birth and death rates
birth_death_rates_by_year = df.groupby('Year')[['Birth rate, crude (per 1,000 people)', 'Death rate, crude (per 1,000 people)']].mean().reset_index()

# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='value', hue='variable', data=pd.melt(birth_death_rates_by_year, id_vars='Year'), palette='muted')

plt.title('Comparison of Birth and Death Rates Over the Years')
plt.ylabel('Rate (per 1,000 people)')
plt.xlabel('Year')
plt.legend(title='Rate Type', loc='upper right')

plt.show()
```

## Electric power consumption 

```python
# Group the data by 'Year' and calculate the max of Electric power consumption
electric_power_by_year = df.groupby('Year')['Electric power consumption (kWh per capita)'].max().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(electric_power_by_year['Year'], electric_power_by_year['Electric power consumption (kWh per capita)'], marker='o', linestyle='-')

plt.title('Electric Power Consumption Over the Years')
plt.xlabel('Year')
plt.ylabel('Electric Power Consumption (kWh per capita)')
plt.grid(True)

plt.show()
```

```python
#  Plotting the line chart for electric power consumption per region over Years
sns.lineplot(x='Year', y='Electric power consumption (kWh per capita)', hue='Region', data=df, marker='o')

plt.title('Electric Power Consumption Over the Years by Region')
plt.xlabel('Year')
plt.ylabel('Electric Power Consumption (kWh per capita)')
plt.legend(title='Region', loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```
